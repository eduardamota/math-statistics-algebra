```python
from random import randint 
aposta = 5
n = 1000000
p = 0
for n in range(1,n):
    G = 0
    y = randint(1,6) #Valor obtido pelo jogador 
    x = randint(1,6) #Valor obtido pelo jogador 
    if (y>x): #Se o ponto do jogador for maior do que o do oponente 
        G = 2*(y-x)
    if (G>aposta):
        p +=1
p = p/n

print(f"A aposta foi de: {aposta}\nValor Torico  de: 1/6\nA chanece de lucro  de {p}")
```
