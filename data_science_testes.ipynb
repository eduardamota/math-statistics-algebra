Temos um p-valor muito pequeno, ou seja, descartamos a hipÃ³tese nula de que a mÃ©dia Ã© igual a 3.42

```python
np.random.seed(20)
temp = notas1.sample(frac=1).rating
def calcula_teste(i):
  media = temp[0:i].mean()
  stat,p = ztest(temp[0:i],value=3.420368675682846)
  return (i,media,p)
medias = np.array([calcula_teste(i) for i in range(2, len(temp))])
medias
plt.plot(medias[:,0],medias[:,1])
plt.plot(medias[:,0],medias[:,2])
plt.hlines(y=0.05,xmin=2,xmax=len(temp), colors='r')
```

```python
#Verificando se os meus dados seguem ums distribuiÃ§Ã£o normal
from scipy.stats import normaltest

stat,p = normaltest(notas1.rating)
print(p)
```

Vemos que o teste nÃ£o segue a distribuiÃ§Ã£o normal, logo, nesse caso vamos utilizar testes nÃ£o paramÃ©tricos

```python
#Teste Wilconx
from scipy.stats import ranksums
_,p = ranksums(notas1.rating,notas593.rating)
p
```

No teste Wilconx temos que na hipÃ³tese nula as duas amoastra tem a mesma distribuiÃ§Ã£o, jÃ¡ a hipÃ³tese nula nos diz que os valores de uma amostra sÃ£o maiores que os de outra amostra.
